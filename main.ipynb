### Different Graph

```python
df1=df.head(4)

x=df1['Job Title']
y=df1['Stipend']

# Line Graph

plt.plot(x, y, marker='o', color='blue', label='Line')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar Graph

plt.bar(x, y, color='green')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
